## Pyplot API vs Object-Oriented API
* Pyplot API  -> Quickly
* OOP API -> Advanced (phuc tap)

## Pyplot API

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color = "green", linestyle = "dashed", label = "sin(x)");
plt.plot(x, np.cos(x), color = "blue", label = "cos(x)")

plt.title("Sin(x) & Cos(x)") # dat titel cho do thi
plt.xlabel("X") # dat ten cho trc x
plt.ylabel("F(x)") # dat ten cho truc y

#plt.xlim([0,4]) # gioi han truc x theo khoang mong muon
#plt.ylim([- 0.6, 2]) # gioi ham truc y theo khoang mong muon
#plt.axis([0,4, -0.7, 0.6])#plt.axis([xmin, xmax, ymin, ymax]) gioi han
plt.axis("tight") # kht lun
plt.axis("equal") # truc x, y bang nhau
plt.legend(); #call ham legend ra moi co dc
```

##  Object-Oriented API

```python
#1 prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize = (5,5)) #figure seze = width and hight of the plot
# fig laf hinh tong
# ax la hinh nhor

#3. plot the data
ax.plot(x, y);

ax.set(title = "A simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis");
```

```python
# OO API
# Prepare New Data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100) # chon mau
sizes = 1000*rng.rand(100) # kich co cua cac cham tron

plt.rcParams["axes.grid"] = False #khong hien thi cai warning
fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s = sizes, c = colors, cmap = "viridis", alpha = 0.3);
fig.colorbar(img1);
```

```python
student_height[:10]
```

### subplots
Multiple plots on one figure link: https://matplotlib.org/3.1.1/gallery/recipes/create_subplots.html

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))
#plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x*2)

ax2.scatter(np.random.rand(10), np.random.rand(10));

ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax4.hist(student_height,bins = 30, color = "green");
```

```python
# option 2:
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                      figsize = (10,5))
#plot data to each axis
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2)

ax[0,1].scatter(np.random.rand(10), np.random.rand(10));

ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax[1,1].hist(student_height,bins = 30, color = "green");
```

# Project: Data Visualization - Calufornia Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Read file

```python
cities = pd.read_csv("./Matplotlib/california_cities.csv")
```

```python
cities.head()
```

```python
# Extract latd = v , longd = Kinh 

lat, lon = cities["latd"], cities["longd"]

population, area = cities["population_total"],cities["area_total_km2"]
```

```python
plt.rcParams['axes.grid'] = False

plt.figure(figsize = (8,6))

#plot using Pyplot API
#plt.scatter(lon,lat, c = "blue",cmap = "viridis", alpha = 0.3);

plt.scatter(lon, lat, 
            c = np.log10(population), 
            linewidths = 0, alpha = 0.5);

plt.axis("equal"); # nhn cho n u
plt.xlabel("longtitude");
plt.ylabel("latitude");
plt.colorbar(label = "log$_{10}$(population)"); # cng thc log10 colorbar
```
